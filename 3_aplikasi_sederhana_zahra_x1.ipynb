#Aplikasi Menghitung Volume Kerucut

```python

jari_jari = int(input("Masukkan jari jari dari kerucut (cm)"))
tinggi = int(input("Masukkan tinggi dari kerucut (cm)"))
phi = 3.14

volume = (1/3) * phi * jari_jari ** 2 * tinggi

print(f"\nVolume dari kerucutnya adalah {volume} cm³")
```
